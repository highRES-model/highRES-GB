```python
py.offline.init_notebook_mode(connected=True) #imports plotly.js to the notebook

DATApath = r'C:\Users\ucqba01\Documents\Local Data\Round 6'
fcost = 100
db = 'hR_m_2002_waves10000_RPS90_fcost%s_newfuelcost.db'%fcost
print('database',db)
print('fcost',fcost)
con = sq.connect(DATApath + '\\' + db)
path = DATApath

VREsupply = getSupplyCurves(fcost, [80, 100, 120, 140], con)

def plotSupplyCurves(df, X, Y, series, title, file, png=False):
    data = []
    df.replace(to_replace='Windoffshore_Mid',value='Windoffshore Mid',inplace=True)
    df.replace(to_replace='Windoffshore_Shallow',value='Windoffshore Shallow',inplace=True)
    seriesList = ['Solar','Windonshore','Windoffshore Shallow','Windoffshore Mid',
                  'Windoffshore Floating 80','Windoffshore Floating 100',
                  'Windoffshore Floating 120','Windoffshore Floating 140']
    for ser in seriesList:
        vredf = makeSupplyCurveDf(df, series, ser, Y, X)
        vredf['cummulative']=vredf['cummulative'].shift(1)
        vredf['cummulative'].fillna(0,inplace=True)
        trace = lineTrace(vredf, 'cummulative', Y, ser)
        data.append(trace)

    yax = dict(title='LCOE (/MWh)',showgrid=True, zeroline=True, showline=True, range=[0,200])
    xax = dict(title='Cumulative Supply (TWh/yr)',showgrid=True, zeroline=True, showline=True)
    

    layout = go.Layout(title=title, xaxis=xax, yaxis=yax)#, legend=dict(x=0.7, y=0.95))
    layout['xaxis']['range']=[0,2000]
    fig = go.Figure(data=data, layout=layout)

    py.offline.iplot(fig,filename='test')
    #py.offline.plot(fig, filename=file)
    
    return data

graphFile = r'C:\Users\ucqba01\Google Drive\Extended Research\Writing\Images\supply curves.html'
supplyCurves = plotSupplyCurves(VREsupply, 'potential_generation_TWh', 'potential_lcoe(/MWh)', 'series', 'Input Supply Curve',
                           graphFile)


```
